#### From the above it can be observed that the columns 'Cut','Color','Clarity' is categorical column so lets convert it into numerical column

```python
# Manulaly mapping and converting the Categorical column into numerical column's

# For Cut column

df['cut'].value_counts()
```

```python
mapping = {'Fair': 4, 'Good': 3, 'Very Good': 2, 'Premium': 1, 'Ideal': 0}
df1 = df.copy()
df1['cut'] = df1['cut'].map(mapping)
```

```python
df1.head(5)
```

```python
df1['cut'].value_counts()
```

```python
# For Clarity column

df1['clarity'].value_counts()
```

```python
mapping = {'SI1': 0, 'VS2': 1, 'SI2': 2, 'VS1': 3, 'VVS2': 4,'VVS1':5 ,'IF':6,'I1':7}
df2 = df1.copy()
df2['clarity'] = df2['clarity'].map(mapping)
```

```python
df2.head(5)
```

```python
df2['clarity'].value_counts()
```

```python
# For Color column

df2['color'].value_counts()
```

```python
mapping = {'G': 0, 'E': 1, 'F': 2, 'H': 3, 'D': 4,'I':5 ,'J':6}
df3 = df2.copy()
df3['color'] = df3['color'].map(mapping)
```

```python
df3.head(5)
```

```python
df3['color'].value_counts()
```

```python
df3.head(15)
```

```python
df3.tail(15)
```

#### It can be observed from above that i was able to succesffully convert the categorical column into numerical column
